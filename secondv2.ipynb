The Second Execise consists on add Latitude and Longitude values on the Wikipedia table

```python
import pandas as pd
```

```python
base = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
data = pd.DataFrame(base[0])
```

```python
data = data[data['Borough'] != 'Not assigned']
data
```

until here we've done on the last execise, then we'll use the file that has latitude an longitude informations

```python
data_latlong =pd.read_csv('Geospatial_Coordinates (1).csv')
```

```python
data_latlong
```

we're going to merge both columns using the Postal code Column, so the latitude and longitude columns from the data_latlong table will be added on the end of the data table

```python
data = pd.merge(data,data_latlong,how='inner',on='Postal Code')
```
