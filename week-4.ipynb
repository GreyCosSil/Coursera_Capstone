# 5. Filtering some columns and cleaning then

```python
# filter columns
filtered_columns = ['venue.name', 'venue.categories'] + [col for col in dataIreland.columns if col.startswith('venue.location.')] + ['venue.id']
dataframe_filtered = dataIreland.loc[:, filtered_columns]


def get_category_name(row):
    try:
        categories_list = row['venue.categories']
    except:
        categories_list = row['venue.name']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

def get_category_id(row):
    try:
        categories_list = row['venue.categories']
    except:
        categories_list = row['venue.name']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['id']
    
# filter the category for each row
dataframe_filtered['id_type'] = dataframe_filtered.apply(get_category_id, axis=1)
dataframe_filtered['name_type'] = dataframe_filtered.apply(get_category_name, axis=1)
    
# clean columns
dataframe_filtered.columns = [col.split('.')[-1] for col in dataframe_filtered.columns]
    
# filter the category for each row

dataframe_filtered.head(10)
```

# 6. Transforming the JSON file into a DataFrame

```python
dataframe_filtered=pd.DataFrame(dataframe_filtered)
```

# 7. Plotting the data in a Map

```python
dublin_map = folium.Map(location=[latitude, longitude], zoom_start=12) # generate map centred around Ecco


# add Ecco as a red circle mark
folium.CircleMarker(
    [latitude, longitude],
    radius=10,
    popup='Ecco',
    fill=True,
    color='Green',
    fill_color='Green',
    fill_opacity=0.6
    ).add_to(dublin_map)


# add popular spots to the map as blue circle markers
for lat, lng, label in zip(dataframe_filtered['lat'], dataframe_filtered['lng'], dataframe_filtered['categories']):
    label = '{}, {}'.format(dataframe_filtered['name_type'], dataframe_filtered['distance'])
    label = folium.Popup(label)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        fill=True,
        color='blue',
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(dublin_map) 

# display map
dublin_map
```
